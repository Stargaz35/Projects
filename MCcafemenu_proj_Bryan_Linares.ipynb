```python
print("Welcome to MC Cafe!")

# Menu display based on user input
resp = input("Would you like to see the menu?(y/n) ")
if resp == "y":
    menu = {
        "Coffee": {
            "Espresso": 1.50,
            "Americano": 2.00,
            "Caffe latte": 2.25,
            "Mocha": 2.25,
            "Cappucino": 2.50,
            "Macchiato": 2.50
        },
        "Tea": {
            "White tea": 1.50,
            "Green tea": 2.00,
            "Oolong tea": 2.25,
            "Black tea": 2.25,
            "Earl Grey": 2.50,
            "Matcha": 2.50
        },
        "Pastries": {
            "Croissant": 5.50,
            "Cinnamon roll": 4.50,
            "Banana bread": 4.25,
            "Pumpkin bread": 3.75,
            "Sticky bun": 4.25,
            "Apple turnover": 3.75
        }
        
    }
    cat1 = list(menu.keys())          # 3 main categories of coffee, tea, and pastries
    cat2 = list(menu.values())      # list of dictionary within dictionary of menu variable
    e = list(cat2[0].keys())          # choices under Coffee
    w = list(cat2[1].keys())          # choices under Tea
    u = list(cat2[2].keys())          # choices under Pastries
    e2 = list(cat2[0].values())     # list of prices of items under Coffee
    w2 = list(cat2[1].values())     #  list of prices of items under Tea 
    u2 = list(cat2[2].values())     # list of prices of items under Pastries
    print(cat1[0].center(40))
    for k in range(len(e)):
        print(f"\t{e[k]}.........${e2[k]:.2f}")
    
    
    print(cat1[1].center(40))
    for d in range(len(w)):
        print(f"\t{w[d]}.........${w2[d]:.2f}")
    
    print(cat1[2].center(40))
    for q in range(len(u)):
        print(f"\t{u[q]}.......${u2[q]:.2f}")

# Item selection from the menu
    resp2 = input("Are you ready to order?(y/n) ")
    if resp2 == "y":
        r = True
        order = {}
        while r:
            sel = input("Enter an item from the menu: ")   
            if not (sel in e or sel in w or sel in u):
                print("Either the item you have entered is not on the menu or you have misspelled it. Please try again")
                continue
            qt = int(input("How many? "))
            order[sel] = qt
            sel_ag = input("Would you like to make another selection?(y/n) ")
            if sel_ag == "n":
                r = False

# Generated order receipt with item selection history
    print("Your Order\n".center(60))
    dr1 = list(order.keys())
    dr2 = list(order.values())
    z1, z2, z3, sub_tot = 0, 0 , 0, 0
    for i in range(len(dr1)):
        if dr1[i] in e:
            ab = e.index(dr1[i])
            z1 = e2[ab]*dr2[i]
            print(f"{dr1[i]}  QTY: {dr2[i]}---------------${z1:.2f}".rjust(50))
            sub_tot += z1
        elif dr1[i] in w:
            cf = w.index(dr1[i])
            z2 = w2[cf]*dr2[i]
            print(f"{dr1[i]}  QTY: {dr2[i]}---------------${z2:.2f}".rjust(50))
            sub_tot += z2
        elif dr1[i] in u:
            rt = u.index(dr1[i])
            z3 = u2[rt]*dr2[i]
            print(f"{dr1[i]}  QTY: {dr2[i]}---------------${z3:.2f}".rjust(50))
            sub_tot += z3
        else:
            pass

# Calculated total amount due in dollars($) from purchased items
    tax_rt = 6/100
    sale_tax = sub_tot * tax_rt
    Total = sub_tot + sale_tax
    print()
    print(f"******SUBTOTAL: ${sub_tot:.2f}".rjust(50))
    print(f"SALES TAX @{tax_rt:.2%}: ${sale_tax:.2f}".rjust(50))
    print(f"******TOTAL: ${Total:.2f}".rjust(50))
    print()
    print("Thank you for visiting MC Cafe!".center(55)) 
        
else:
     print("Thank you for visiting MC Cafe!")
```
